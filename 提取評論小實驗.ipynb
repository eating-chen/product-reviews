```python
# 載入停用詞
def read_txt_list(filepath):  
    words = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return words

stop_words_file_path = "/Users/eating/Desktop/eating_django/eating_django/et_project/nlp_explore/jieba_dict/stop_words.txt"

stop_list = read_txt_list(stop_words_file_path)
```

```python
# 第一步：pos 保留名詞及英文
product_feature = [] # 保存每句的名詞及英文
all_sentence = [] # 將每句斷詞後的詞保存下來，這樣jieba只要做一次
all_sentence_pos = [] # 將每句斷詞後的詞性保存下來，這樣jieba只要做一次
for line in all_t:
    temp = []
    each_line = []
    each_line_pos = []
    seg_line = pos.cut(line)
    for w in seg_line:
        word = w.word
        flag = w.flag
        if (str(flag).lower() == 'n' or str(flag).lower() == 'eng'):
            temp.append(word)
        if word not in stop_list and word != '' and str(flag).lower() != 'x':
            each_line.append(word)
            each_line_pos.append(flag)
    product_feature.append(temp)
    all_sentence.append(each_line)
    all_sentence_pos.append(each_line_pos)

for p_f in product_feature:
    print(p_f)
```

```python
# 第二步：apriori 保留常見的名詞及英文，作為產品特徵
# 轉成pandas的型式
def deal(data):
    return data.dropna().tolist()

product_feature_pd = pd.DataFrame(product_feature)
product_feature_arr = product_feature_pd.apply(deal,axis=1).tolist()
te = TransactionEncoder() # 定義模型
product_feature_tf = te.fit_transform(product_feature_arr)
product_feature_df = pd.DataFrame(product_feature_tf ,columns=te.columns_)
# 跑 apriori , support 設 0.01
frequent_itemsets = apriori(product_feature_df,min_support=0.01, use_colnames=True) # use_colnames=True表示使用元素名字，預設的False使用列名代表元素
frequent_itemsets.sort_values(by='support',ascending=False,inplace=True)
f = frequent_itemsets[frequent_itemsets.itemsets.apply(lambda x: len(x) >= 1 and len(x) < 4)]

# 將型態轉成list以方便後面做處理
itemset=f['itemsets'].tolist()
for i in range(len(itemset)):
    itemset[i] = list(itemset[i])
    print(itemset[i])
```

```python
# 第三步：特徵裁剪
# 去除多餘的特徵，採用paper上的方法 compact phrase, 針對長度>=2的特徵集合使用
for item in range(len(itemset)):
    if(len(itemset[item]) > 1):
        feature_list = itemset[item]
        item_count = 0 # 紀錄這個特徵集出現的句子數
        for msg in all_sentence:
            count_compute = [-999 for i in range(len(feature_list))] #記錄用，用來記錄每個產品特徵的位置
            for j in range(len(feature_list)):
                try:
                    count_compute[j] = msg.index(feature_list[j])
                except:
                    count_compute[j] = -999
            count_compute.sort()
            # 長度2的情況，這邊照paper設定距離為3
            if len(feature_list) == 2 and -999 not in count_compute and abs(count_compute[0]-count_compute[1]) <= 3: # 紀錄是否特徵都有出現在這句中
                item_count += 1
            # 長度3的情況
            elif len(feature_list) == 3 and -999 not in count_compute and (abs(count_compute[2]-count_compute[1]) <= 3 and abs(count_compute[1]-count_compute[0]) <= 3):
                item_count += 1
        if item_count < 2: #若句子數小於2就把該特徵設為'r1'
            itemset[item] = 'r1'

itemset = [x for x in itemset if x != 'r1'] #保留不是'r1'的產品特徵

# 採剪長度為 1、2 的特徵集合
p_support = 3 
record = [0 for i in range(len(itemset))] # 紀錄這比資料是否尋訪過
for item in range(len(itemset)):
    if(len(itemset[item]) < 3):
        feature_list = itemset[item]
        super_set=[]
        item_count = 0 # 紀錄這個特徵集出現的句子數
        for set_item in itemset:
            if set(feature_list).issubset(set(set_item)): # 先尋訪是否為超集
                # 把差集的部分取出來
                temp_list = list(set(set_item) - set(feature_list))
                for t in temp_list:
                    super_set.append(t)
        for msg in all_sentence:
            count_compute = [-1 for i in range(len(feature_list))]
            for j in range(len(feature_list)):
                count = 0
                for i in range(0, len(msg)):
                    if(feature_list[j] == msg[i]):
                        count_compute[j] = count
                        break
                    count += 1
            superset_compute = [-1 for i in range(len(super_set))]
            for j in range(len(super_set)):
                for i in range(0, len(msg)):
                    if(super_set[j] == msg[i]):
                        superset_compute[j] = 1
                        break
            if -1 not in count_compute and 1 not in superset_compute:
                item_count += 1
        if item_count < p_support: #若出現句子數小於p_support，也設為'r1'
            itemset[item] = 'r1'

itemset = [x for x in itemset if x != 'r1']
for i_s in itemset:
    print(i_s)
```

```python
# 第四步：意見詞提取，結巴的意見詞主要是以'vi'為主
feature_opinion_pair = [] # 取得{feature, opinion} pair
opinion_list=[]
min_value = 9999
# 提取特徵-意見pair
for item in range(len(itemset)):
    feature_list = itemset[item]
    neighbor_opinion=[]
    for msg in range(len(all_sentence)):
        count_compute = 0
        temp_adj = []
        msg_list = []
        adj_position = []
        for i in range(len(all_sentence[msg])):
            msg_list.append(all_sentence[msg][i])
            current_word = all_sentence[msg][i]
            current_pos = all_sentence_pos[msg][i]
            if str(current_pos).lower() == 'vi':
                temp_adj.append(current_word)
                adj_position.append(i)
        if set(feature_list).issubset(set(msg_list)):
            count_compute = 1      
        if count_compute == 1:
            # find feature position
            feature_count=[-1 for i in range(len(feature_list))]
            for i in range(len(feature_list)):
                feature_count[i] = msg_list.index(feature_list[i])

            min_position = min(feature_count)
            max_position = max(feature_count)
            for i in range(len(adj_position)):
                if(abs(min_position - adj_position[i]) < min_value):
                    neighbor_opinion.append(temp_adj[i])
                if(abs(max_position - adj_position[i]) < min_value):
                    neighbor_opinion.append(temp_adj[i])
    neighbor_opinion = list(set(neighbor_opinion))
    if len(neighbor_opinion) != 0:
        feature_opinion_pair.append({
            'feature':feature_list,
            'opinion':neighbor_opinion
        })
for f in feature_opinion_pair:
    print(f)
```

```python
# 第五步：找出評論句
# 檢查每個句子，若有產品特徵及其評論詞，則加入comment_list中
comment_list = []
feature_comment_pair = []
for feature in feature_opinion_pair:
    temp_comment_list = []
    # 先看 feature 有沒有再這句中
    for msg in all_sentence:
        feature_count = 0
        if set(feature['feature']).issubset(set(msg)):
            feature_count = 1
        # 若有feature就看意見詞存不存在
        if feature_count == 1:
            for i in range(len(feature['opinion'])):
                if feature['opinion'][i] in msg:
                    opinion_index = msg.index(feature['opinion'][i])
                    check_pos = all_sentence_pos[all_sentence.index(msg)][opinion_index]
                    if str(check_pos).lower() == 'vi':
                        comment_list.append(all_t[all_sentence.index(msg)])

comment_list = list(set(comment_list))
count = 1
for c in comment_list:
    print("第", count, "句：", c)
    count += 1
```
